```julia
using PyPlot
```

```julia
k = 3 # degree of spline
n = 9 # number of control points
m = n + k + 1 # number of knots in T
P = [0 0 4 0 0 2 5 0 0 ; 0 1 1 -1 0 2 1 0 1]
P
```

```julia
# knots
T = [0 0 0 0 1.0/6 2.0/6 3.0/6 4.0/6 5.0/6 1 1 1 1]'
```

```julia
t = 0.9
N = Dict()
N[1] = 1.0*(t .>= T[1:end-1]) .* (t .< T[2:end])
```

```julia
invd = Dict()
for i = 1:k+1
    invd[i] = T[(i+1):end] - T[1:(end-(i))]
    
    # take care of empty knot spans
    for j = 1:size(invd[i])[1]
        if invd[i][j] != 0
            invd[i][j] = 1.0/invd[i][j]
        end
    end
end

for i = 1:k+1
    if i > 1
        N[i] = (t - T[1:end-i]) .* (invd[i-1].*N[i-1])[1:end-1] + (T[i+1:end] - t) .*(invd[i-1].*N[i-1])[2:end]
    end
end

N
```

```julia
tt = linspace(0,1,101)
rx = zeros(size(tt))
ry = zeros(size(tt))

for j = 1:size(tt)[1]
    t = tt[j]
    N[1] = 1.0*(t .>= T[1:end-1]) .* (t .< T[2:end])
    
    for i = 1:k+1
        if i > 1
            N[i] = (t - T[1:end-i]) .* (invd[i-1].*N[i-1])[1:end-1] + (T[i+1:end] - t) .*(invd[i-1].*N[i-1])[2:end]
        end
    end
    
    rx[j] = (P[1,:]'*N[k+1])[1]
    ry[j] = (P[2,:]'*N[k+1])[1]
end
```

```julia
figure()
plot(rx,ry)
show()
```
